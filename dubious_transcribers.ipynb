```python
import csv

with open('./tables/linguistic_data.csv', 'r') as f:
    reader = csv.DictReader(f)
    document_ling_data = [row for row in reader]
document_ling_data
```

```python
[row['ASL'] for row in document_ling_data if row['Document Name'].startswith('1001')]
```

```python
from utils.document import TextDocument
from utils.data.raw import text_files_dir


data = []
for row in document_ling_data:
    name = row['Document Name']
    set_num = name[0]
    path = f"{text_files_dir}/set0{set_num}/{name}"
    doc = TextDocument(path)
    num_periods = doc.content('Participant').count('.')

    data.append({'Document': name, 
                 'Set' : set_num, 
                 'Transcript' : doc.transcript_number, 
                 'No. periods' : num_periods, 
                 'ASL' : float(row['ASL'])})
```

```python
sorted(data, key=lambda x: x['No. periods'], reverse=True)
```
