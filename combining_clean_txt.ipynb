This code combines the cleaned text_files into individual interviews. The first number in the file IDs is the set that the interview belongs to. The next three numbers indicate the interview number.  E.g. 2##1_### indicates the file belongs to Interview 1 of Set 2 (clinicians).

```python
from pathlib import Path
from collections import defaultdict
```

```python
CLEAND_DIR = Path('./data/mathews/documents/text_files/')
COMBO_DIR = Path('./data/mathews/documents/combined_interviews/')
```

```python
def merge_files_by_prefix(source_dir, dest_dir):
    files_by_prefix = defaultdict(list)

    problem_files = []
    
    # Ensure destination directory exists
    if not dest_dir.exists():
        dest_dir.mkdir()

    # Group files by prefix
    for textfile in source_dir.rglob('*.txt'):
        interviewID = textfile.name[:4]
        files_by_prefix[interviewID].append(textfile)
    print(len(files_by_prefix.items()))
    # Merge files for each prefix
    for interviewID, filenames in files_by_prefix.items():
        # Define the name of the merged file
        merged_filepath = dest_dir/interviewID
        # Open the destination file in write mode ('w')
        with open(merged_filepath.with_suffix(".txt"), 'w') as outfile:
            # Sort the filenames to ensure consistent order of content
            for filename in sorted(filenames):
                #infilepath = source_dir/filename
                with open(filename, 'r', encoding='utf8') as infile:
                    try:
                        outfile.write(infile.read())
                    except (UnicodeDecodeError, UnicodeEncodeError):
                        problem_files.append(filename.name)

        print(f"Merged {len(filenames)} files into {merged_filepath}")
    with open(Path('./data/unicode_problems.txt'), 'w') as log:
        log.write('\n'.join(problem_files))
```

```python
merge_files_by_prefix(CLEAND_DIR,COMBO_DIR,)
```
