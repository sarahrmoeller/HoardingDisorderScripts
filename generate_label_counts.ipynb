```python
import os

REVIEWS_DIR = "./data/HD_set1_1-7/REVIEW/"
# Get all .json files in REVIEW dir
json_files = [f for f in os.listdir(REVIEWS_DIR) if os.path.splitext(f)[1] == '.json']
test_file = json_files[0]
hoarder_flag = int(test_file[0] == '0')
```

```python
import json

with open(REVIEWS_DIR + test_file) as f:
    raw_data = json.load(f)

data = raw_data['data']
data.keys()
```

```python
data['project'], data['document'], data['kinds']
```

```python
data['labelSets'], data['labelerInfo']
# Save all label names to a list
LABELS = {label_item['labelName'] for label_item in data['labelSets'][0]['labelItems']}
LABELS
```

```python
data
```

```python
# This is where the meat is at
label_data = data['spanLabels']
label_data
```

```python
from collections import Counter

cntDict = Counter([label['labelItem']['labelName'] for label in label_data])
for label in set(LABELS).difference(cntDict.keys()):
    cntDict[label] = 0
cntDict
cntDict['Total'] = sum(cntDict.values())
```

```python
row = {
    'filename' : data['document']['name'], ''
    'isHoarder' : hoarder_flag, 
    **cntDict
}
```
