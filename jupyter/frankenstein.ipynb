```python
import json
import os

projects = os.listdir('./data/')
projects.remove('.gitignore')
projects.sort()
```

```python
files = []
for project in projects:
    review_dir = f'./data/{project}/REVIEW/'
    project_files = [f for f in os.listdir(review_dir) if os.path.splitext(f)[1] == '.json']
    files.extend(project_files)
files.sort()
```

```python
joined_file_data = []
# i.e. 001_001.txt.json -> (split) [001, 001.txt.json]
prev_interview_number = int(files[0].split('_')[0])
last_index = 0
for curr_index in range(1, len(files)):
    file = files[curr_index]
    curr_interview_number = int(file.split('_')[0])
    if curr_interview_number > prev_interview_number:
        joined_file_data.append(files[last_index:curr_index])
        last_index = curr_index
        prev_interview_number = curr_interview_number
joined_file_data.append(files[last_index:]) # loop doesn't append the last list of files, so we do it here
```

```python
# sanity check
all(
    file_data[0].split('_')[0] == file.split('_')[0] for file_data in joined_file_data for file in file_data
)
```
