## Test: Comparing Agreement on a Particular Label between Two Authors

For this test, we'll choose two particular authors and one particular label that we want to see how they agree on.

First, we'll see which documents they've mutually annotated. We'll assign `None` to the Kappa value if no documents were mutually annotated.

```python
import utils.datasaur as data

current_label = 'Clarification'
author_pair = ('sotodaniel', 'savannah_cherry')

common_docs = [doc.name for doc in data.by_doc
               if all(doc.name in map(lambda d: d.name, author_docs[author]) 
                      for author in author_pair)]
common_docs
```

Next, we'll filter the label dictionary for labels that only come from these documents.

```python
filtered_dict = {
    author : label_list for author, label_list in author_labels.items()
    if author in author_pair
}
common_docs = set.intersection(*[set(author_labels[author].keys()) 
                                 for author in author_pair])
common_docs
```

With this, we can now begin what's needed. To calculate Cohen's Kappa, we're 
going to use the `cohen_kappa_score` from the `scikit-learn` package. From the
[documentation](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.cohen_kappa_score.html),
we need a list of labels from the first annotator, `y1`, and a list of labels
from the second annotator, `y2` (order doesn't matter).

There's sort of an issue here in defining what a label is, because in Cohen's 
Kappa, it's assumed that a label is held in common between both annotators. 
In our context, if this assumption were true, this would mean that every single 
incomplete clause in the data has been at least identified to annotators, and 
annotators had the choice to give these pre-defined spans a certain label. 
However, that's not really the case here: annotators were also given the choice
on which spans were and were not an incomplete clause, and *then* chose to give
their defined label a label categorization.

To remedy this, I am going to define what a label as the following:

**Definition** (label). A span that was annotated by at least one of the two 
annotators in question.

I chose this definition carefully to make sure that that *both* annotators could 
have annotated each label, with the possiblity of all kinds of mutual decisions (true
positives/negatives and false positives/negatives).

```python
def sameish_span(text_position1: dict, text_position2: dict, 
                 token_window=3) -> bool:
    """
    Check if the position of two spans are the same. We say same(ish) since
    we allow for a window of +/- `token_window` tokens.
    """
    if text_position1['start']['tokenIndex'] - token_window < 0:
        text_position1['start']['tokenIndex']
    start_ok = abs(text_position1['start']['tokenIndex'] - 
                   text_position2['start']['tokenIndex']) <= token_window
    end_ok = abs(text_position1['end']['tokenIndex'] - 
                 text_position2['end']['tokenIndex']) <= token_window
    # In the case when the row numbers are the same, the check is pretty 
    # simple. Just make sure the start and end token indices are within the 
    # `token_window.
    if text_position1['row'] == text_position2['row']:
        return start_ok and end_ok
   # Otherwise, it gets kind of complicated.
    elif text_position1['row'] - text_position2['row'] == 1:
        ...
    elif text_position2['row'] - text_position1['row'] == 1:
        ...
    return False
```
